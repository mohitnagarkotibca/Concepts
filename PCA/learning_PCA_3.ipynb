```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Requesting a dataset with 500 examples of 5 features
plt.style.use('fivethirtyeight')
mc= make_blobs(n_samples=500,n_features=5,centers=2,cluster_std=2,random_state=2)
df=pd.DataFrame(mc[0])
df
```

```python
#features
x0=mc[0][:,0]
x1=mc[0][:,1]
x2=mc[0][:,2]
x3=mc[0][:,3]
x4=mc[0][:,4]

#labels
z= mc[1]
plt.scatter(x= x0,y= x1,c=z,edgecolors='black',label='X0')
plt.xlabel('X1')
plt.ylabel('Y')
```
