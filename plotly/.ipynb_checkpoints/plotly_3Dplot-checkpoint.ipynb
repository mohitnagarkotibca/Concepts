```python
import pandas as pd
```

```python
data= pd.read_csv('pokemon.csv')
```

```python
data.columns
```

```python
import plotly.figure_factory as ff
x=data['Sp. Atk']

y=data['Total']
fig= ff.create_2d_density(x,y,)
fig.update_layout(
    xaxis_title='speed',
    yaxis_title='attack'
)
fig.show()

```

```python

import plotly.graph_objects as go
```

```python
x= data['Attack']
y= data['Defense']
z= data['Speed']

fig=go.Figure(layout={'margin':{'l':0,'r':0,'b':0,'t':0}})
fig.add_trace(
                go.Scatter3d(
                x= x,
                    y=y,
                    z=z,
                    mode='markers',
                        marker=dict(
                    size=12,
                    color='#ff0000',                # set color to an array/list of desired values
                    colorscale='Viridis',   # choose a colorscale
                    opacity=0.5
                )
                )
)
```
