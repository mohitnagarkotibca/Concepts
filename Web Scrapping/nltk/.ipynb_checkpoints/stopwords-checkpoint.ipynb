```python
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
text=nltk.Text(nltk.word_tokenize(text))
```

```python
tokenized= sent_tokenize(text)
word_li=[]
for token in tokenized:
    words= word_tokenize(token)
    [word_li.append(word) for word in words]
    
```

```python
stop_words= nltk.corpus.stopwords.words('english')
for word in ['the','if','and','was','to']:
    stop_words.append(word)
print(len(stop_words))
stop_words

dist=nltk.FreqDist(text)
# dist.plot(stop_words=stop_words)

dist=nltk.FreqDist(dict((word,freq) for word,freq in dist.items() if word not in ['the','In','as','The','of','in','and','was','on','to','a','which','their'] and word.isalpha()))
dist.plot(20)

```
