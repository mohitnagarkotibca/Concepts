```python
import tensorflow as tf
import skimage.io as sk
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=sk.imread('data/plain_cup.png',as_gray=True)
plt.imshow(image,cmap='gray')

# kernel_h = tf.constant([ [1,2,1], [0,0,0], [-1,-2,-1] ],shape=(3,3,1,1),dtype='float32')
# kernel_v = tf.constant([ [1,0,1], [2,0,-2], [-1,0,-1] ],shape=(3,3,1,1),dtype='float32')

kernel_h = tf.constant([ [0,-1,0], [-1,4,-1], [0,-1,0] ],shape=(3,3,1,1),dtype='float32')
kernel_v = tf.constant([ [-1,-1,-1], [-1,8,-1], [-1,-1,-1] ],shape=(3,3,1,1),dtype='float32')


tensor_image= tf.constant(image,shape=(1,image.shape[0],image.shape[1],1),dtype='float32')

plt.imshow(tensor_image.numpy()[0])  

h= tf.nn.conv2d(input= tensor_image,filters=kernel_h,strides=[1,1,1,1],padding='SAME')[0]
v= tf.nn.conv2d(input= tensor_image,filters=kernel_v,strides=[1,1,1,1],padding='SAME')[0]

fig,ax= plt.subplots(nrows=2,ncols=2,figsize=(10,8))
# ax[0][0].imshow(h)
# ax[0][0].title.set_text('Horizontal edge detection')
# ax[0][1].imshow(v)
# ax[0][1].title.set_text('vertical edge detection')
# ax[1][0].imshow(h,alpha=0.5)
# ax[1][0].imshow(v)
# ax[1][0].title.set_text('Blend of horz on vert with 0.5')
# ax[1][1].imshow(h)
# ax[1][1].imshow(v,alpha=0.5)
# ax[1][1].title.set_text('Blend of vert on horz with 0.5')

ax[0][0].imshow(tf.nn.relu(h))
ax[0][1].imshow(tf.nn.leaky_relu(h))
ax[1][0].imshow(tf.nn.sigmoid(h))
ax[1][1].imshow(tf.nn.relu6(h))
```
