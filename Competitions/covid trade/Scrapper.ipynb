```python
with open('data/country_code/countrycode.json','r') as file:
    li= file.read()
country_ids= json.loads(li[3:])
country_code= [d['id'] for d in country_ids['results']]
for year in [2020,2021]:
    for code in country_code:
        print(year,":",code)
        res= requests.get(f'https://comtrade.un.org/api/get/plus?r=699&ps={year}&p={code}&rg=ALL&cc=TOTAL&max=1000&type=C&freq=M&head=H')
        li= json.loads(res.content)
        if li['dataset'] != []:
            json_object= json.dumps(li)
            with open(f"data/India_with_{code}_code_country_{year}.json",'w') as file:
                file.write(json_object)
```
